```r
#libraries
library(ggplot2)
library(dplyr)
data(mtcars)
```

```r
#What is a good graph for showing two numeric variables?
# a scatterplot
ggplot(mtcars, aes(x=drat, y=hp)) +
  geom_point()
```

```r
#What is a good graph for 1 categorical and one numeric variable?
#Summarise the numeric variable and make a bar graph
mtcars %>% group_by(gear) %>%
  summarise(hp_median =median(hp)) %>%
ggplot(aes(x=gear, y=hp_median)) +
  geom_bar(stat="identity")
```

```r
#What is a good graph for two categorical variables?
#a bar graph
mtcars$gear <- as.factor(mtcars$gear)
mtcars$cyl <- as.factor(mtcars$cyl)
ggplot(mtcars, aes(x=gear, fill = cyl)) +
  geom_bar()
#or side by side
ggplot(mtcars, aes(x=gear, fill = cyl)) +
  geom_bar(position="dodge")
```

```r
#What is a good graph for two numeric and 1 categorical variable?
#A fancy scattergraph

mtcars$gear <- as.factor(mtcars$gear)
mtcars$cyl <- as.factor(mtcars$cyl)

ggplot(mtcars, aes(x=hp, y=drat, colour=cyl, shape=cyl)) +
  geom_point(alpha=0.5)
# or facets
ggplot(mtcars, aes(x=hp, y=drat, colour=cyl)) +
  geom_point(alpha=0.3) + facet_wrap( ~ cyl, ncol=2) 
```

```r
#What is a good graph for three categorical variables?
#Bar graph with facets

mtcars$gear <- as.factor(mtcars$gear)
mtcars$cyl <- as.factor(mtcars$cyl)
mtcars$vs <- as.factor(mtcars$vs)
ggplot(mtcars, aes(x=gear, fill = vs)) +
  geom_bar(position="dodge") + facet_wrap( ~ cyl, ncol=2) 
```

```r
#How to avoid overplotting?
#Overplotting is where having points directly on top of each other 
#makes it hard to see how many are there

ggplot(mtcars, aes(x=drat, y=gear)) +
  geom_point()
#use see through points
ggplot(mtcars, aes(x=drat, y=gear)) +
  geom_point(alpha=0.2)
#and/or jittering
ggplot(mtcars, aes(x=drat, y=gear))  +
  geom_point(alpha=0.4, position=position_jitter(width=.1,height=.1))
```

```r
#How to get rid of the grey background?
#switch to black and white

ggplot(mtcars, aes(x=drat, y=gear)) +
  geom_point() + theme_bw()
```
