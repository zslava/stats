## R basic commands

```r
### matrix
x=matrix(data=c(1,2,3,4), nrow=2, ncol=2)
x
```

```r
matrix(c(1,2,3,4),2,2,byrow=TRUE)
```

```r
sqrt(x)
x^2
```

```r
### rnorm() to generate a vector of random normal variables
x=rnorm(50)
y=x+rnorm(50,mean=50,sd=.01) ## specify n points  + mean  +sd
par(mfrow=c(2,2))
plot(x)
plot(y)
paste("correlation x,y:",cor(x,y))
```

```r
### to reproduce exact set of random variables, use set.seed()
set.seed(1303)
rnorm (50)
```

```r
##mean and var and sd
set.seed(3)
y=rnorm(100)
mean(y)
var(y)
sqrt(var(y))
sd(y)
```

### Graphics

```r
x=rnorm(100)
y=rnorm(100)
plot(x,y,xlab="this is the x-axis",ylab="this is the y-axis",main="Plot of X vs Y")
```

```r
## to save a chart in a file
pdf("Figure.pdf")
plot(x,y,col="green")
dev.off() ## we are done creating a plot
```

```r
### sequences 
x=seq(1,10)
x
x=1:10  # this is a syntax for range
x
```

```r
##Contour plots
x=seq(-pi,pi,length=50)
y=x
f=outer(x,y,function(x,y)cos(y)/(1+x^2))
contour(x,y,f)
contour(x,y,f,nlevels=45,add=T)
```

```r
fa=(f-t(f))/2
contour(x,y,fa,nlevels=15)
```

```r
## image function makes a heatmap
image(x,y,fa)
```

```r
persp(x,y,fa)
```

```r
persp(x,y,fa,theta=30,phi=20) #turn projection
```
