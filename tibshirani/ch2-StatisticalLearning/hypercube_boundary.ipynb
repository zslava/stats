### compute a proportion of boundary points in a hypercube

Defined as points which have at least in 1 dimension either the 5% smallest value  or  5% biggest value


```r
## first try a normal 3d cube  N=3
```

```r
set.seed(11)
npp<-10000 #number of points per plane
d<-3
```

```r
cube<-data.frame( matrix(runif(npp*d, min=0,max=1), ncol=d) )
```

```r
## get the max and min for each row
cube[, "max"] <- apply(cube, MARGIN = 1, max) 
cube[, "min"] <- apply(cube, MARGIN = 1, min) 
#see the added "max" and "min" columns
head(cube,2)
```

```r
# set boundary to 1 if min is less than .05 
cube$boundary <-0
cube$boundary[cube$min < .05] <-1
##  or max greater than .95
cube$boundary[cube$max > .95] <-1
### see the new column boundary
head(cube)
```

```r
nrow(cube[cube$boundary==1,])
#expected answer diffence between all cube and non-boudnary internal cube (it has lenth 0.9)
(1-.9^d)
```

```r
### hypercube with d = 50
```

```r
set.seed(11)
npp<-10000
d<-50
```

```r
hcube<-data.frame( matrix(runif(npp*d, min=0,max=1), ncol=d) )
hcube[, "max"] <- apply(hcube, MARGIN = 1, max) 
hcube[, "min"] <- apply(hcube, MARGIN = 1, min)
hcube$boundary <-0
hcube$boundary[hcube$min < .05] <-1
##  or max greater than .95
hcube$boundary[hcube$max > .95] <-1
### see the new column boundary
t(head(hcube,10))
```

```r
dim(hcube)
```

```r
#in 10000 rows  the number of rows which has at least 1 point in a vector of lenght 50
#in a boundary zone i.e. all those rows are attributed to the boundary zone
nrow(hcube[hcube$boundary==1,])
```

```r
nrow(hcube[hcube$boundary==1,]) / npp
```
