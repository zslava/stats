#### Housing Values in Suburbs of Boston
This data frame contains the following columns:
* crim: per capita crime rate by town (or suburb)
* zn: proportion of residential land zoned for lots over 25,000 sq.ft.
* indus: proportion of non-retail business acres per town.
* chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
* nox: nitrogen oxides concentration (parts per 10 million).
* rm: average number of rooms per dwelling.
* age: proportion of owner-occupied units built prior to 1940.
* dis: weighted mean of distances to five Boston employment centres.
* rad: index of accessibility to radial highways.
* tax: full-value property-tax rate per \$10,000.
* ptratio: pupil-teacher ratio by town.
* black: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
* lstat: lower status of the population (percent).
* medv: median value of owner-occupied homes in \$1000s.

```r
## 10.b
pairs(Boston, cex=0.1,col=2)
```

```r
#10.c
par(mfrow=c(2,3))
plot(medv~crim, data=Boston, cex=0.3) #home price inversely related to crime rate
plot(medv~lstat, data=Boston, cex=0.3, col=2) #price inverseley related to a higher percentage of lower status folks
plot(medv~rm, data=Boston,cex=0.3, col=3) #price proportional to a number of rooms
plot(medv~zn, data=Boston,cex=0.3,col=4) #price somewhat proportional to the proportion of residential land zones
boxplot(medv~chas, data=Boston, cex=0.3,col=5, xlab="charles River") #with water view price is higher
plot(medv~dis, data=Boston, cex=0.3,col=6) #price somewhat proportional to a higher distance from employment centers
```

```r
#10.d
par(mfrow=c(2,2))
plot(medv~tax,data=Boston) #higher price home in lower tax rates areas
plot(medv~ptratio,data=Boston) #price tends to be higher with higher with higher number of pupils per teacher
plot(medv~black,data=Boston) # high proportion of black population corresponds to lower prices 
plot(medv~age,data=Boston)
```
